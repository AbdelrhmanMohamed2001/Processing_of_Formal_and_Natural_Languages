```python
# IMPLEMENTATION OF DFA

x= int(input("Enter no. of states: "))
state=[input("Enter states: ") for i in range(0,x)]
y= int(input("Enter no. of alphabets: "))
alphabets=[input("Enter alphabets: ") for i in range(0,y)]
last=input("Enter accepted states: ")
k=[0 for i in range(len(state))]
for i in range (len(state)):
  k[i]=[0 for j in range(len(alphabets))]
  for j in range(len(alphabets)):
    k[i][j]= input("from "+ state[i]+ " if "+ alphabets[j]+ " go : ")

def spot(q,w):
  lis.append(k[state.index(q)][alphabets.index(w)])
  return k[state.index(q)][alphabets.index(w)]
```

```python
def DFA_count(input_string):
    current_state = 0
    count = 0
    if(len(input_string) < 5):
         return 0
    else:
       
     for i in range(len(input_string)):
        if current_state == 0 and input_string[i] == '1':
            current_state = 1
        elif current_state == 0 and input_string[i] == '0':
            current_state = 0
        elif current_state == 1 and input_string[i] == '0':
            current_state = 2
        elif current_state == 1 and input_string[i] == '1':
            current_state = 1
        elif current_state == 2 and input_string[i] == '1':
            current_state = 3
        elif current_state == 2 and input_string[i] == '0':
            current_state = 0
        elif current_state == 3 and input_string[i] == '0':
            current_state = 4
        elif current_state == 3 and input_string[i] == '1':
            current_state = 1
        elif current_state == 4 and input_string[i] == '1':
            count += 1
            if input_string[i-1:i+1] == '10101':
                current_state = 1
            else:
                current_state = 3
        elif current_state == 4 and input_string[i] == '0':
            current_state = 4
        else:
            current_state = 0

    return count

```

```python
transitions = {
    0: {'0': [0], '1': [0, 1]},
    1: {'0': [2], '1': [0, 1]},
    2: {'0': [0], '1': [0, 3]},
    3: {'0': [4], '1': [0, 1]},
    4: {'0': [0], '1': [0, 1, 3]}
}

accept_state = 4

def nfa_count(input_string):
  current_states = [0]
  count = 0
  if(len(input_string) < 5):
         return 0
  else:
   for symbol in input_string:
        next_states = []
        for state in current_states:
            if symbol in transitions[state]:
                next_states.extend(transitions[state][symbol])
        
        current_states = next_states
        
        if accept_state in current_states:
            count += 1
            
  return count
```
