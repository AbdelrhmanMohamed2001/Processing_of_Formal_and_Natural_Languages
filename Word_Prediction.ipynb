```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
import os
import glob
import numpy as np
import gensim.downloader as api
import tensorflow as tf

# Define the data processing functions
def load_data(folder_path):
    data = []
    files = glob.glob(os.path.join(folder_path, "*.txt"))
    for file in files:
        with open(file, "r") as f:
            text = f.read()
            data.append(text)
    return data

def tokenize(text):
    tokens = text.split()
    return tokens

def preprocess(tokens):
    stop_words = stopwords.words('english')
    punctuation = [",", ".", "!", "?", ";", ":", "-", "--", "(", ")", "[", "]", "{", "}", "'", "\""]
    tokens = [token.lower() for token in tokens if token.lower() not in stop_words and token not in punctuation]
    return tokens
```

```python
# Load the dataset
data_folder = "/content/enronsent1"
data = load_data(data_folder)

# Tokenize and preprocess the data
tokenized_data = [tokenize(text) for text in data]
preprocessed_data = [preprocess(tokens) for tokens in tokenized_data]
print("length of preprocessed_data : ", len(preprocessed_data))
preprocessed_data
```

```python
# Split the data into training and validation sets
train_data = preprocessed_data[0:10]
val_data = preprocessed_data[11:15]
print("length of train_data",len(train_data))
print("length of val_data",len(val_data))
train_data
```

```python
# Download the glove-wiki-gigaword-50 model
model = api.load("glove-wiki-gigaword-50")
```

```python
# Define the word prediction functions
def predict_next_word(prefix):
    words = prefix.split()
    if len(words) == 0:
        return ""
    if len(words) == 1:
        return model.most_similar(words[0])[0][0]
    last_word = words[-1]
    last_two_words = " ".join(words[-2:])
    if last_two_words in model:
        similar_words = model.most_similar(last_two_words)
        for word, score in similar_words:
            if word != last_word:
                return word
    else:
        similar_words = model.most_similar(last_word)
        for word, score in similar_words:
            if word != last_word:
                return word
    return ""
```

```python
def check_prediction(prediction):
    response = input(f"Is your next word: '{prediction}'? ")
    if response.lower() == "yes":
        return prediction
    else:
        correct_word = input("Enter the correct word: ")
        if correct_word == "-1":
            return ""
        else:
            return correct_word

def generate_sentence():
    next_word = input("Enter next word (-1 to terminate): ")
    sentence = next_word
    while True:
        if next_word == "-1":
            break
        prefix = f" {next_word}"
        prediction = predict_next_word(prefix)
        next_word = check_prediction(prediction)
        if next_word == "":
            break
        sentence += " " + next_word
        
    return sentence.strip()
```

```python
# Example 1
final_sentence_1 = generate_sentence()
print(f"Your final Sentence is '{final_sentence_1}'")
```
